```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
iris = sns.load_dataset("iris")
iris.head(5)
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
sns.set()
%matplotlib inline
```

```python
sns.swarmplot(x="species", y="petal_length", data=iris)
```

```python
sns.swarmplot(x="species", y="petal_width", data=iris)
```

```python
sns.swarmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.swarmplot(x="total_bill", y="sex", data=tips)
```

```python
sns.stripplot(x="total_bill", y="sex", data=tips)
```

```python
sns.set()
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
sns.set_style("whitegrid")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_style("ticks")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_style("white")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_style("dark")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_style("ticks")
sns.boxplot(x = "day", y = "total_bill", data = tips)
sns.despine()
```

```python
sns.set_style("ticks")
sns.boxplot(x = "day", y = "total_bill", data = tips)
sns.despine(left=True)
```
